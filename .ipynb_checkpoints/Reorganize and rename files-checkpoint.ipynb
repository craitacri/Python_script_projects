```python
base_dir="C:\\Users\\coman\\Documents\\GitHub\\Python_script_projects\\Foderprojec"
slash='\\'
parent=[]
for folder , sub_folders , files in os.walk("machine make 2",):
    print("Currently looking at folder: "+ folder)
    print('\n')
    print("THE SUBFOLDERS ARE: ")
    if "\\" in folder:
        for sub_fold in sub_folders:
            print("\t Subfolder: "+sub_fold )
            #print("PRINT1!!!"+str(sub_fold)+"-")
            for parents in parent:
                # creating parent and directories in parent in new format
                directories=folder.split('\\')
                if directories[0] in parents and directories[1] in parents:
                                         
                    os.chdir(base_dir+slash+parents)
                    #print('changed inside directory')
                    #os.getcwd()
                    os.mkdir("T-"+sub_fold)
                    lastcreateddir=os.getcwd()+'\\'+"T-"+sub_fold
                    #print (lastcreateddir)
                    #print('created new inside directory t-')
                    os.chdir(base_dir)
                    #print('changed outside directory')
                    #os.getcwd()
                    #print('\n')
                else:
                    pass
    else:
        for sub_fold in sub_folders:
            print("\t Subfolder: "+sub_fold )
            #print("PRINT2!!!"+str(sub_fold)+"-"+str(folder))
            os.mkdir("TO-"+sub_fold+"-"+str(folder))
            parent.append("TO-"+sub_fold+"-"+str(folder))
        print('\n')
    print("THE FILES ARE: ")
    for f in files:
        print("\t File: "+f)
        shutil.move(folder+"\\"+f,lastcreateddir)
        #shutil.move(f,lastcreateddir)
    print('\n')
print("Files moved!")
```
